```python
import datetime, requests, yfinance
from alpaca_trade_api.rest import REST
from getpass import getpass

BASE_URL = "https://paper-api.alpaca.markets"
API_KEY = getpass('Enter your API KEY: ')
SECRET_KEY = getpass('Enter your SECRET KEY: ')

api = REST(key_id=API_KEY, secret_key=SECRET_KEY, base_url=BASE_URL)
```

```python
news = api.get_news('TSLA')

news
```

```python
news = api.get_news('LULU', start='2022-04-19', end='2022-04-19', include_content=True)
news
```

```python
for story in news:
  print(f'新聞標題: {story.headline}')
  print(f'新聞內容: {story.summary}')
  for image in story.images:
    print(image['url'])
```

```python
from transformers import pipeline

classifier = pipeline('sentiment-analysis')
```

```python
classifier(' but sad')
```

```python
for story in news:
  label = classifier(story.summary)[0]['label']
  if label == 'NEGATIVE':
    label = '負面'
  else:
    label = '正面'
  score = classifier(story.summary)[0]['score']
  score = score * 100

  print(f'新聞標題: {story.headline}\n')
  print(f'情緒指數: {label}, 分數為: {score}\n\n\n')
  print(f'新聞內容: {story.summary}\n\n\n')
  
```
